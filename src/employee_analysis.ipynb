```python
from datetime import datetime
transactions["day"] = pd.to_datetime(transactions["day"])
transactions
```

```python
def week_of_year(row):
    week_nr = row["day"].isocalendar()[2]
    return week_nr

transactions["day of week"] = transactions.apply(week_of_year, axis=1)
```

```python
import plotly.express as px
dayofweek = transactions.groupby(["nday","day of week"]).count()["customer"]
dayofweek
```

```python
days = [dayofweek.index[i][0] for i in range(len(dayofweek))]
weekdays = [dayofweek.index[i][1] for i in range(len(dayofweek))]
df_weekday = pd.DataFrame({"days":days, "day of week":weekdays, "transactions":dayofweek.values})
df_weekday
```

```python
fig = px.box(df_weekday, x="day of week", y="transactions", color="day of week", title="Distribution of transactions per day of week")
fig.show()
```

```python
transactions_hourly = transactions.groupby(["nday","hour"]).agg({"customer":"nunique"})
days = [transactions_hourly.index[i][0] for i in range(len(transactions_hourly))]
hours = [transactions_hourly.index[i][1] for i in range(len(transactions_hourly))]
customer_values = [transactions_hourly.iloc[i][0] for i in range(len(transactions_hourly))]

df_transactions = pd.DataFrame({"days":days, "hours":hours, "customers":customer_values})
df_transactions
```

```python
fig = px.box(df_transactions, x="hours", y="customers")
fig.show()
```

```python
transactions_hourly = transactions.groupby(["nday","day of week"]).agg({"customer":"nunique"})
days = [transactions_hourly.index[i][0] for i in range(len(transactions_hourly))]
hours = [transactions_hourly.index[i][1] for i in range(len(transactions_hourly))]
customer_values = [transactions_hourly.iloc[i][0] for i in range(len(transactions_hourly))]

df_transactions = pd.DataFrame({"days":days, "day of week":hours, "customers":customer_values})
df_transactions
```

```python
fig = px.box(df_transactions, x="day of week", y="customers", title="Distribution of customers per day of week")
fig.show()
```
